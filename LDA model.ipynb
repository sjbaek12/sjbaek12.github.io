```python
from gensim.corpora.dictionary import Dictionary
from gensim import corpora
import codecs
```

```python
from konlpy.tag import Kkma
import gensim
k = Kkma()
```

```python
t1 = codecs.open('t1.txt', 'r', encoding='utf-8')
t2 = codecs.open('t2.txt', 'r', encoding='utf-8')
```

```python
t1_lines = t1.readlines()
t2_lines = t2.readlines()
```

```python
t1 = ''
for t in t1_lines:
    t1 += t
    
t2 = ''
for t in t2_lines:
    t2 += t
```

```python
t2
```

```python
t1_k = k.nouns(t1)
t2_k = k.nouns(t2)
```

```python
t1_k
```

```python
stopword = ["차", "활", "용해", "글", "일", "고", "수", "문", "반", "명", "만", "누구", "당시", "문", "셈", "식", "강병원", "라", "문제", "인", "우리", "민국", "대한", "민국", "의원", "말", "이후"]
```

```python
for word in t1_k:
    if word in stopword:
        t1_k.remove(word)
        
for word in t2_k:
    if word in stopword:
        t2_k.remove(word)
```

```python
text = [t1_k, t2_k]
```

```python
dictionary = corpora.Dictionary(text)
```

```python
print(dictionary.token2id)
```

```python
corpus = [dictionary.doc2bow(text) for text in text]
```

```python
corpus
```

```python
lda = gensim.models.ldamodel.LdaModel(corpus=corpus, id2word=dictionary, num_topics=3)
```

```python
lda.print_topics(3)
```
