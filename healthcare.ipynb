```python
import codecs
f = codecs.open("SV health.txt", 'r',encoding='utf8')
```

```python
lines=f.readlines()
for line in lines:
    print(line)
```

```python
company_list = []
dic = {}
for line in lines:
    if "####" in line:
        dic = {}
        dic["Name"] = line
        continue
    elif "Founded:" in line:
        dic["Founded"] = line
        continue
    elif "Mission Statement" in line:
        dic["Mission"] = line 
        continue
    elif "Description:" in line:
        dic["Description"] = line
        continue
    elif "Website:" in line:
        dic["Web"] = line
        continue
    elif "---------------" in line:
        company_list.append(dic)
        dic={}
        continue
```

```python
company_list
```

```python
company_founded = []
for company in company_list:
    if "Founded" in company.keys():
        company_founded.append(company)
```

```python
year_list = []
for company in company_founded:
    if "2009" in company["Founded"] :
        year_list.append(company)
year_list
```
